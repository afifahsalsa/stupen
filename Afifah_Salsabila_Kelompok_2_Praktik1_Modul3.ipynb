# 6. Lakukan Pembaharuan

**a. Data pada variabel pack yang memiliki category_name CREAM LIQUEURS diubah menjadi 40**

```python
#cek data dengan category_name cream liqueurs
query = """ SELECT pack, category_name FROM sales WHERE category_name = 'CREAM LIQUEURS'; """
pd.read_sql_query(query, con)
```

```python
#update variabel pack menjadi 40
query = """ UPDATE sales SET pack = 40 WHERE category_name = 'CREAM LIQUEURS' """
con.execute(query)
```

```python
#cek datanya lagi
query = """ SELECT pack, category_name FROM sales WHERE category_name = 'CREAM LIQUEURS'; """
pd.read_sql_query(query, con)
```
